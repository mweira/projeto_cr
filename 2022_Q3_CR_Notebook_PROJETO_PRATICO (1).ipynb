Em seguida, podemos visualizar o grafo

```python
plt.figure(figsize =(30, 30))
node_color = [G.degree(v) for v in G]

nx.draw_networkx(G,
                 node_color = node_color, 
                 alpha = 0.7,
                 node_size=500,
                 with_labels = True,
                 font_weight='bold',
                 cmap = plt.cm.Blues)

plt.axis('off')
plt.tight_layout();
```

# **Análise e Resultados**



```python
print('Número de vértices: ', G.number_of_nodes())
print('Número de arestas: ', G.number_of_edges())
degrees = [val for (node, val) in G.degree()]
print('Grau médio: ', "{:.2f}".format( round(sum(degrees) / G.number_of_nodes(), 2)) )
print('Grau máximo: ', max(degrees))
print('Densidade: ', "{:.2f}".format( nx.density(G)) )
print('Coeficiente de agrupamento médio: ', "{:.2f}".format( nx.average_clustering(G)) )
print('Transitividade: ', "{:.2f}".format( round(nx.transitivity(G), 2)) )
print('Conectividade: ', 'Conexo' if nx.is_connected(G) else "Não é possível calcular")
print('Componentes conexos: ', nx.number_connected_components(G))
print('Distância média: ', "{:.2f}".format( nx.average_shortest_path_length(G) ) if nx.is_connected(G) else "infinito")
print('Diâmetro: ', nx.diameter(G) if nx.is_connected(G) else "Não é possível calcular")
```

# **Graus**

**Distribuição de Graus**

```python
degrees = [G.degree(n) for n in G.nodes()]
av_degree = round(sum(degrees) / G.number_of_nodes(), 2)


plt.hist(degrees, label='Distribuição de Graus')
plt.title("Distribuição de Graus")
plt.axvline(av_degree,color='r',linestyle='dashed',label='Grau Médio')
plt.legend()
plt.ylabel("Contagem")
plt.xlabel("Grau")
plt.show()
```

# **Centralidades**

```python
layout = nx.spring_layout(G, iterations=50)
```


* **Centralidade de grau** 

```python
cor_vertices = [nx.degree_centrality(G)[v] for v in G]
tamanho_vertices = [100 + 100 * G.degree(v) for v in G]

rotulo_vertices = {} 
for v in G.nodes(): 
  rotulo_vertices[v] = round(nx.degree_centrality(G)[v], 2)

plt.figure(figsize=(25, 25))
nx.draw(G,layout, nodelist=G.nodes(), node_size = tamanho_vertices, node_color = cor_vertices, edge_color= 'grey')
nx.draw_networkx_labels(G, layout, rotulo_vertices, font_weight = 'bold', font_color = 'black' );
plt.show()
```

* **Centralidade relativa de grau**


```python
cor_vertices = [nx.degree_centrality(G)[v] for v in G]
tamanho_vertices = [150 + 150 * G.degree(v) for v in G]

rotulo_vertices = {} 
for v in G.nodes(): 
  rotulo_vertices[v] = round(nx.degree_centrality(G)[v], 2)

plt.figure(figsize=(25, 25))
nx.draw(G,layout, nodelist=G.nodes(), node_size = tamanho_vertices, node_color = cor_vertices, edge_color= 'grey')
nx.draw_networkx_labels(G, layout, rotulo_vertices, font_weight = 'bold', font_color = 'black' );
plt.show()
```
